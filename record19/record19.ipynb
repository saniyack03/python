```python
#copy odd lines of one file to another
file1=input("Enter the source file:")
file2=input("Enter the destintion file:")
f1=open(file1,"r")
f2=open(file2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0:
        continue
    f2.write(line)
f1.close()
f2.close()
print("Odd lines copied successfully from",file1,"to",file2)
```

```python
#read each row from a csv file
import csv
fname=input("Enter CSV file name:")
f1=open(fname,"r")
reader=csv.reader(f1)
for row in reader:
    print(row)
f1.close()
```

```python
#read specific columns
import csv
fname=input("Enter the CSV file name:")
with open(fname,"r")as f1:
    reader=csv.DictReader(f1)
    print("\nDepartment Details:")
    print("----------------")
    for row in reader:
        print("Deptid:",row['Deptid'],"\tDepName:",row['DeptName'])
```

```python
#python dictionary to a csv file
data=[
    {"Name":"Saniya","Age":"22","Marks":"85"},
    {"Name":"Nefla","Age":"21","Marks":"90"},
    {"Name":"Aradhya","Age":"21","Marks":"83"}
]
 
    
with open("students_dict.csv", "w", newline='') as file:
    fieldnames = ["Name", "Age", "Marks"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("Dictionary data written successfully!")

with open("students_dict.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
    
```
